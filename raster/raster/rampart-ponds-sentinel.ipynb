```python
from pystac_client import Client

catalog = Client.open("https://earth-search.aws.element84.com/v0")
```

```python
time_range = "2022-01-01/2022-10-29"
area_of_interest = {
    "coordinates": [
        [
            [
                -122.98736687545757,
                49.83337619352599
            ],
            [
                -122.9890708254656,
                49.807043701381104
            ],
            [
                -122.92578125373085,
                49.80688660575089
            ],
            [
                -122.9248887084886,
                49.83400422905237
            ],
            [
                -122.98736687545757,
                49.83337619352599
            ]
        ]
    ],
    "type": "Polygon"
}

search = catalog.search(collections=["sentinel-s2-l2a-cogs"], intersects=area_of_interest, datetime=time_range)
items = search.get_all_items()
len(items)
items[0]
```

```python
import geopandas

items_dict = items.to_dict()
for i in range(len(items)):
    items_dict["features"][i]["properties"]["visual_href"]=items[i].assets["visual"].href

df = geopandas.GeoDataFrame.from_features(items_dict, crs="epsg:4326")
df["visual_href"]
```

```python
import pandas

df['month'] = pandas.DatetimeIndex(df['datetime']).month
df
```

```python
filtered_df = None
for month in range(1, 11):
    items_month = df[(df['month']==month) & (df['sentinel:valid_cloud_cover'])]
    selected_items = items_month[items_month["eo:cloud_cover"] == items_month["eo:cloud_cover"].min()]
    selected_item = selected_items[0:1]
    filtered_df = pandas.concat([filtered_df, selected_item])
```
